```python
import pandas as pd
```

```python
news=pd.read_csv("E:/Downloads/archive/Combined_News_DJIA.csv")
news.head()
```

```python
train=news[news['Date']>'20130627']
```

```python
test=news[news['Date']<'20130627']
```

```python
## Removing Puntuations
data=train.iloc[:,2:25]
data
```

```python
data.replace ("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
## Renaming the columns 
list1=[i for i in range(23)]
new_Index=[str(i) for i in list1]
data.columns=new_Index
data.head(2)
```

```python
## converting the letter into lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
headlines=[]
for row in range (0,len(data.index)):
    headlines.append(' '.join(str(x)for x in data.iloc[row,0:23]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## Implement Bag of words
countvector = CountVectorizer(ngram_range=(3,3))
traindataset=countvector.fit_transform(headlines)
```

```python
## implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for The Dataset
test_transform = []
for row in range (0,len(test.index)):
     test_transform.append(' '.join(str(x)for x in test.iloc[row,2:25]))
test_dataset=countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
## checking accuracy
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score =accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
